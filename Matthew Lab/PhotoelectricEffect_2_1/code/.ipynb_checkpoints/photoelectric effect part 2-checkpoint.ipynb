```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mm4 = np.transpose(np.loadtxt('../data/pt2_run1.csv',delimiter = ','));
mm2 = np.transpose(np.loadtxt('../data/pt2_run2.csv',delimiter = ','));
mm8 = np.transpose(np.loadtxt('../data/pt2_run3.csv',delimiter = ','));
```

```python

figure, subplots = plt.subplots(nrows = 1, ncols = 2,figsize=(10, 5))
plot1, plot2 = subplots
plot1.plot(mm2[0,:],mm2[1,:],'r.',label = "2 mm slit")
plot1.plot(mm4[0,:],mm4[1,:],'b.',label = "4 mm slit")
plot1.plot(mm8[0,:],mm8[1,:],'y.', label = "8 mm slit")
plot1.set_xlabel("Voltage(V)")
plot1.set_ylabel("Current(pA)")
plot1.legend()
plot2.axvline(-0.62, color='yellow')
plot2.axvline(-0.6,color = 'red')
plot2.axvline(0,color = 'blue')
plot1.axvline(-0.62, color='yellow')
plot1.axvline(-0.6,color = 'red')
plot1.axvline(0,color = 'blue')
plot2.plot(mm2[0,15:],mm2[1,15:],'r.',label = "2 mm slit")
plot2.plot(mm4[0,15:],mm4[1,15:],'b.',label = "4 mm slit")
plot2.plot(mm8[0,15:],mm8[1,15:],'y.', label = "8 mm slit")
plot2.set_xlabel("Voltage(V)")
plot2.set_ylabel("Current(pA)")
plot2.legend()
plt.show()
```
